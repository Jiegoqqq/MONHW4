```python
#.ps檔轉成.png檔
from PIL import Image
import os

# 設定輸入和輸出資料夾路徑
input_folder = "/home/jiego/1122MON/HW15/static/images/own_S09_RNAfold"
output_folder = "/home/jiego/1122MON/HW15/static/images/own_S09_RNAfold_PNG"

# 確保輸出資料夾存在，如果不存在則創建
os.makedirs(output_folder, exist_ok=True)

# 列出輸入資料夾中的所有檔案
for filename in os.listdir(input_folder):
    if filename.endswith(".ps"):
        # 處理每個 .ps 檔案
        ps_path = os.path.join(input_folder, filename)
        image = Image.open(ps_path)

        # 將 .ps 檔案轉換為 .png 檔案
        png_filename = os.path.splitext(filename)[0] + ".png"
        png_path = os.path.join(output_folder, png_filename)
        image.save(png_path, "PNG")

        print(f"Converted {filename} to {png_filename}")

print("Conversion completed.")
```
